Importar pandas y numpy

Acceder a los datos
Ingresar al link https://github.com/cienciadedatos/datos-de-miercoles/tree/master/datos/2019/2019-08-07, clic en el archivo felicidad.csv y clic en la el ícono RAW. Copiar el link del navegador y utilizar pandas para importarlo dandole un nombre.

```python
datos_felmun = pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
datos_felmun = datos_felmun.set_index('pais')
datos_felmun
```

Para calcular la medida de tendencia central y dispersión de la felicidad a nivel mundial para el primer y último año incluido en el set de datos, primero se debe averiguar cual es el primer y ultimo año del que se tiene registro, aplicamos la funcion describe a la columna anio y observamos que el valor minimo es 2005 y maximo 2018, en base a esos años calcularemos dichas medidas.

```python
datos_felmun['anio'].describe()
```

Ya sabiendo los años,se procede a aislarlos para poder calcular las medidas de media y desviación estándar que son las elegidas, eso lo haremos mediante el uso de índices y operandos de la siguiente manera:

```python
d2005 = datos_felmun[datos_felmun['anio'] == 2005]
```

```python
d2018 = datos_felmun[datos_felmun['anio'] == 2018]
```

Luego aplicamos la funcion de describe a cada set de datos del año 2005 y 2018

```python
medidas_2005 = d2005.describe()
medidas_2005
```

```python
medidas_2018 = d2018.describe()
medidas_2018
```

Finalmente obtenemos los datos con la función loc y iloc, también añadimos la función T para poder crear una tabla resumen al final, de manera que se pueda comparar mejor los valores de cada país elegido tanto para el 2005 como el 2018, que en el presente caso es Venezuela y España

```python
Venezuela_2005 = pd.DataFrame(d2005.loc['Venezuela'])
Venezuela_2005.T
```

```python
Venezuela_2018 = pd.DataFrame(d2018.loc['Venezuela'])
Venezuela_2018.T
```

```python
España_2005 = pd.DataFrame(d2005.loc['España'])
España_2005.T
```

```python
España_2018 = pd.DataFrame(d2018.loc['España'])
España_2018.T
```

```python
Res_2005 = medidas_2005
Res2005 = Res_2005.iloc[1:3,:]
Res2005
```

```python
Res_2018 = medidas_2018
Res2018 = Res_2018.iloc[1:3,:]
Res2018
```

```python
Resumen = pd.concat([Venezuela_2005.T, España_2005.T, Res2005, Venezuela_2018.T, España_2018.T, Res2018])
Resumen
```

Para el análisis final podemos observar que tanto Venezuela como España estaban sobre la medida mundial con respecto a felicidad en el año 2005 esto puede deberse pricipalmente a sus altos puntajes en soporte social y afectoo positivo, además de bajos puntajes en afecto negativo. No obstante para el año 2018 observamos una baja de la media de felicidad en ambos países y a nivel mundial, si embargo España aún se mantiene sobre la media principalmente por tener un log pib similar que ahora supera la media y alta esperanza de vida respecto al resto del mundo, por otro lado Venezuela bajo de la media, esto debido a disminución del log pib, aumento de la percepción de corrupción, aumento del afecto negativo, disminución de confianza en el gobierno y una desigualdad de salarios reflejada en un aumento importante en el valor GINI.
